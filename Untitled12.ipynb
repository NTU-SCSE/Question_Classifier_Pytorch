```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas_ml import ConfusionMatrix
import seaborn as sns
import numpy as np
import sklearn
```

```python
df_tmp = pd.read_csv("faq_feedback - Sheet1.csv")
```

```python
df_tmp.head(10)
```

```python
def transform(x, y, z):
    if x == y:
        return z
    return x

df_tmp['true_label'] = df_tmp['true_label'].apply(lambda x: transform(x, 'Academic', 'Academics'))
```

```python
df_tmp.head(10)
```

```python
df_tmp['true_label'] = df_tmp['true_label'].apply(lambda x: transform(x, 'Other services', 'Other Services'))
```

```python
%matplotlib inline
```

```python
confusion_matrix = ConfusionMatrix(df_tmp['true_label'], df_tmp['predicted'])
```

```python
confusion_matrix
```

```python
confusion_matrix.print_stats()
```

```python
label_map = ['Academics', 'Admission & Financial Services',  'Campus Life & Accommodation', 'Other Services', 'Outreach & Exchange']
```

```python
s = sklearn.metrics.confusion_matrix(df_tmp.true_label, df_tmp.predicted)
```

```python
df_cm = pd.DataFrame(data = s, columns=label_map, index=label_map)
plt.figure(dpi=100)
        
heatmap = sns.heatmap(df_cm, annot=True, fmt='d')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=70, ha='right', fontsize=5)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=20, ha='right', fontsize=5)
```
